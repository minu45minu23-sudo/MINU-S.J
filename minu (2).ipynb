```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# ---------------------------------
# Read Data from CSV File
# ---------------------------------
df = pd.read_csv('book3.csv')

print("\nOriginal Dataset")
print(df)

# ---------------------------------
# Calculate Total and Average
# ---------------------------------
subjects = ['Math', 'Science', 'English', 'Computer']

df['Total'] = df[subjects].sum(axis=1)
df['Average'] = df[subjects].mean(axis=1)

# ---------------------------------
# Assign Grade
# ---------------------------------
def assign_grade(avg):
    if avg >= 90:
        return 'A+'
    elif avg >= 80:
        return 'A'
    elif avg >= 70:
        return 'B'
    elif avg >= 60:
        return 'C'
    else:
        return 'Fail'

df['Grade'] = df['Average'].apply(assign_grade)

# ---------------------------------
# Pass/Fail Status
# ---------------------------------
df['Status'] = df[subjects].apply(
    lambda x: 'Pass' if all(mark >= 50 for mark in x) else 'Fail',
    axis=1
)

print("\nStudent Performance Report")
print(df)
```

```python
# Subject-wise Performance
# ---------------------------------
subject_avg = df[subjects].mean()

print("\nSubject-wise Average Marks")
print(subject_avg)

# ---------------------------------
# Class-wise Comparison
# ---------------------------------
class_avg = df.groupby('Class')[subjects].mean()

print("\nClass-wise Performance")
print(class_avg)

# ---------------------------------
# Save Processed Results
# ---------------------------------
df.to_csv('student_result_report.csv', index=False)
print("\nProcessed report saved as 'student_result_report.csv'")

# ---------------------------------
# Chart 1: Subject-wise Average
# ---------------------------------
plt.figure(figsize=(7,5))
subject_avg.plot(kind='bar')
plt.title('Subject-wise Average Marks')
plt.ylabel('Marks')
plt.xticks(rotation=0)
plt.show()
```

```python
# Chart 3: Pass vs Fail
# ---------------------------------
status_count = df['Status'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(status_count,
        labels=status_count.index,
        autopct='%1.1f%%')
plt.title('Pass vs Fail Distribution')
plt.show()
```

```python
# Chart 4: Class-wise Comparison
# ---------------------------------
class_avg.plot(kind='bar', figsize=(8,5))
plt.title('Class-wise Subject Comparison')
plt.ylabel('Average Marks')
plt.xticks(rotation=0)
plt.show()
```
